```python
import os
```

```python
def convert_files_to_parquet() -> str:
    
 
        json_data_dir="../finance_artifact/data_ingestion/20260402_143422/downloaded_files"
        data_dir = "../finance_artifact/data_ingestion/feature_store"
        output_file_name = "finance_complaint"
        os.makedirs(data_dir, exist_ok=True)
        print("Current working directory:", os.getcwd())
        print("JSON dir:", json_data_dir)
        print("Exists:", os.path.exists(json_data_dir))
        file_path = os.path.join(data_dir, f"{output_file_name}")
        
        if not os.path.exists(json_data_dir):
            return file_path
        for file_name in os.listdir(json_data_dir):
            json_file_path = os.path.join(json_data_dir, file_name)
            
            df = spark_session.read.json(json_file_path)
            print(df.count())
            if df.count() > 0:
                df.write.mode('append').parquet(file_path)
                print("write success")
        return file_path
     
```
